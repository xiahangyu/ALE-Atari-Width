```python
import tensorflow as tf
import numpy as np
from AEModel import AEModel
import os
import matplotlib.pyplot as plt

ae = AEModel()
sess = tf.Session()
sess.run(tf.global_variables_initializer())
saver = tf.train.Saver()
saver.restore(sess, "./ckpt/model")

def predict(screen_values):
    screen = np.array(screen_values)
    screen = np.reshape(screen, [1, 33600])
    hidden_state = sess.run(ae.z, feed_dict={ae.x: screen})
    hidden_state = np.reshape(hidden_state, [34560])
    return [v for v in hidden_state]

def close():
    sess.close()

n_train_screens = 4444
n_dev_screens = 1
screen_height = 210
screen_width = 160
dev_screens = np.zeros((n_dev_screens, screen_height*screen_width))
screen_dir = "./screens/freeway/"


def loadData(dir):
    for i in range(n_dev_screens):
        path = dir + str(n_train_screens+i+1) + ".matrix"
        with open(path, "r") as f:
            pixels = f.read().split(' ')[:-1]
            pixels = list(map(int, pixels))
            dev_screens[i] = np.array(pixels)
loadData(screen_dir)
hidden_state = np.array(predict(dev_screens[0]))
#dev_screens = np.reshape(dev_screens, [33600])
# #diff = dev_screens-recon
# for v in range(33600):
#     print(v,":",diff[v],dev_screens[v],recon[v])
#print(np.min(hidden_state))
for v in range(33600):
    if hidden_state[v] > 10 or hidden_state[v] < -10:
        print(v,":",hidden_state[v])


'''
fig, axs = plt.subplots(2, 1, figsize=(1, 2))
for example_i in range(1):
    axs[0][example_i].imshow(
        np.reshape(dev_screens[example_i, :], (210, 160)))
    axs[1][example_i].imshow(
        np.reshape(np.reshape(recon, (210,160))))
        
fig.show()
plt.draw()
plt.waitforbuttonpress
'''
```
